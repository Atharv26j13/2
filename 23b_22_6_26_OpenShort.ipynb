```python
import numpy as np
import pandas as pd

import cv2
import matplotlib.pyplot as plt

emptyImg = np.zeros(shape=(512,512,3),dtype=np.int16)

cv2.rectangle(emptyImg,pt1=(380,10),pt2=(500,150),color=(0,255,0),thickness=10)
cv2.circle(emptyImg,center=(250,250),radius=40,color=(0,0,255),thickness=-1)
cv2.line(emptyImg,pt1=(100,250),pt2=(300,400),color=(255,0,0),thickness=3)
plt.imshow(emptyImg)
```
