```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_csv('gapminder(2007).csv')
data.head()
```

```python
data.isnull().any()
```

```python
data.info()
```

```python
sns.set_style('darkgrid')
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('dark')
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('white')
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('ticks')
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.countplot(x=data['continent'],palette='rocket')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.countplot(x=data['continent'],color='darkorchid')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('paper',font_scale=1.5,rc={'figure.figsize':(10,6)})
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('poster',font_scale=0.5,rc={'figure.figsize':(10,6)})
sns.countplot(x=data['continent'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_context('paper',font_scale=1.5,rc={'figure.figsize':(10,6)})
sns.countplot(x=data['continent'])
plt.yticks(rotation=90)
plt.show()
```

```python
#EVERYTHING

sns.set_style('darkgrid')
sns.set_context('paper',font_scale=1,rc={'figure.figsize':(10,6)})
sns.countplot(x=data['continent'],palette='rocket')
sns.despine(left=True,bottom=True)
plt.yticks(rotation=90)
plt.show()
```
