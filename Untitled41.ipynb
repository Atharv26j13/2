```python
import random
import matplotlib.pyplot as plt


a = []

b = []
c = []
d = []
for i in range(100):
    if random.randint(1,3) == 1:
        a.append(random.randint(1,100))
    elif random.randint(1,2) == 1:
        a.append(random.randint(-100,0))
    else:
        a.append(chr(random.randint(ord('A'), ord('Z'))))

print(a)
for i in a:
  if str(i) == i:
    b.append(i)
  elif i>0:
    c.append(i)
  else:
    d.append(i)
print("\n",b,"\n",c,"\n",d)

bx = []
cx = []
dx = []

for i in b:
  if i == "A":
    bx.append("1")
  elif i == "Z":
    bx.append("100")
  elif i == "B":
    bx.append("1")
  elif i == "Y":
    bx.append("100")
  elif i == "C":
    bx.append("1")
  elif i == "X":
    bx.append("100")
  elif i == "D":
    bx.append("1")
  elif i == "W":
    bx.append("100")
  elif i == "E":
    bx.append("1")
  elif i == "V":
    bx.append("100")
  elif i == "F":
    bx.append("1")
  elif i == "U":
    bx.append("100")
  elif i == "G":
    bx.append("1")
  elif i == "T":
    bx.append("100")
  elif i == "H":
    bx.append("1")
  elif i == "S":
    bx.append("100")
  elif i == "I":
    bx.append("1")
  elif i == "R":
    bx.append("100")
  elif i == "J":
    bx.append("1")
  elif i == "Q":
    bx.append("100")
  elif i == "K":
    bx.append("1")
  elif i == "P":
    bx.append("100")
  elif i == "L":
    bx.append("1")
  elif i == "O":
    bx.append("100")
  elif i == "M":
    bx.append("1")
  else:
    bx.append("100")
print(bx)

for i in c:
  if i > 50:
    cx.append("100")
  else:
    cx.append("1")
print(cx)

for i in d:
  if i > -50:
    dx.append("100")
  else:
    dx.append("1")
print(dx)

th1 = []
th2 = []
th3 = []

for i in range(len(bx)):
  th1.append(i)

for i in range(len(cx)):
  th2.append(i)

for i in range(len(dx)):
  th3.append(i)

plt.scatter(th1,bx)
plt.scatter(th1,b)
plt.plot(th1,b)
plt.show()

plt.scatter(th2,cx)
plt.scatter(th2,c)
plt.plot(th2,c)
plt.show()

plt.scatter(th3,dx)
plt.scatter(th3,d)
plt.plot(th3,d)
plt.show()
```
