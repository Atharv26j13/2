```python
import matplotlib.pyplot as plt
import seaborn as sns
import random as r
```

```python
from numpy import size
years=[]
for i in range(10):
  years.append(i+2009)

apple=[r.randint(50,100)]
microsoft=[r.randint(50,100)]
samsung=[r.randint(50,100)]
xiaomi=[r.randint(50,100)]
LG=[r.randint(50,100)]
android=[r.randint(50,100)]

for i in range(9):
  apple.append(apple[-1]+r.randint(-10,10))
  microsoft.append(microsoft[-1]+r.randint(-10,10))
  samsung.append(samsung[-1]+r.randint(-10,10))
  xiaomi.append(xiaomi[-1]+r.randint(-10,10))
  LG.append(LG[-1]+r.randint(-10,10))
  android.append(android[-1]+r.randint(-10,10))

sns.set_theme('paper')

plt.plot(years,apple,label="Apple",color='red')
plt.plot(years,microsoft,label="Microsoft",color='blue')
plt.plot(years,samsung,label="Samsung",color='green')
plt.plot(years,xiaomi,label="Xiaomi",color='purple')
plt.plot(years,LG,label="LG",color='orange')
plt.plot(years,android,label="Android",color='lime')
plt.legend()

plt.title("Stock Prices",size=20)
plt.xlabel("Years",size=15)
plt.ylabel("Prices",size=15)

plt.fill_between(years, apple, label="Apple Area", color='red', alpha=0.3)
plt.fill_between(years, microsoft, label="Microsoft Area", color='blue', alpha=0.3)
plt.fill_between(years, samsung, label="Samsung Area", color='green', alpha=0.3)
plt.fill_between(years, xiaomi, label="Xiaomi Area", color='purple', alpha=0.3)
plt.fill_between(years, LG, label="LG Area", color='orange', alpha=0.3)
plt.fill_between(years, android, label="Android Area", color='lime', alpha=0.3)

plt.annotate(f'  Apple:{apple[-1]-apple[1]}',(years[-1],10))
plt.annotate(f'  Microsoft:{microsoft[-1]-microsoft[1]}',(years[-1],20))
plt.annotate(f'  Samsung:{samsung[-1]-samsung[1]}',(years[-1],30))
plt.annotate(f'  Xiaomi:{xiaomi[-1]-xiaomi[1]}',(years[-1],40))
plt.annotate(f'  LG:{LG[-1]-LG[1]}',(years[-1],50))
plt.annotate(f'  Android:{android[-1]-android[1]}',(years[-1],60))

plt.show()
```
