```python
classes=[]
x = ''
y = ''
z = ''
legendary = ''
ancient = ''
for i in gen1_pokemon:
  if len(i['type']) == 1:
    x = 'Single Type'
  else:
    x = 'Multi Type'
  if i['hp'] > 70:
    y = 'strong'
  else:
    y = 'weak'
  if i['damage'] > 70:
    z = 'Damager'
  else:
    z = 'Defender'
  if i['name'] == 'Mewtwo' or i['name'] == 'Mew' or i['name'] == 'Moltres' or i['name'] == 'Dragonite' or i['name'] == 'Zapdos' or i['name'] == 'Articuno':
    legendary = True
  else:
    legendary = False
  if i['name'] == 'Omanyte' or i['name'] == 'Kabuto' or i['name'] == 'Omastar' or i['name'] == 'Kabutops':
    ancient = True
  else:
    ancient = False

  classes.append({'type':[x,z],'health':y,'legendary':legendary,'ancient':ancient})
  print(f'name: {i["name"]}\ntype: {x} {z}\nlegendary: {legendary}\nancient: {ancient}\n')
```

```python
xl = []
yl = []
nl = []
for i in classes:
  xl.append(i['legendary'])
  yl.append(i['ancient'])
for i in gen1_pokemon:
  nl.append(i['name'])
print(nl)
print(xl)
print(yl)
import matplotlib.pyplot as plt
plt.bar(nl,yl)
plt.bar(nl,xl)
plt.xlabel('Pokemon')
plt.ylabel('Legendary/Ancient')
plt.title('Legendary/Ancient Pokemon')
plt.show()
```
