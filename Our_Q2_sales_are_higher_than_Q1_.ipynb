```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Titanic Dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
q1 = data['Age'].quantile(0.25)
q2 = data['Age'].quantile(0.50)
q3 = data['Age'].quantile(0.75)
print (q1, q2, q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
plt.hist(data['Age'])
plt.ylabel('Count of passengers')
plt.xlabel('Age')
plt.show()
```

```python
q1 = data['Fare'].quantile(0.25)
q2 = data['Fare'].quantile(0.50)
q3 = data['Fare'].quantile(0.75)
print (q1, q2, q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
plt.hist(data['Fare'])
plt.ylabel('Count of passengers')
plt.xlabel('Fare')
plt.show()
```

```python
plt.boxplot(data['Age'])
plt.title("Age Box Plot")
plt.show()
```
