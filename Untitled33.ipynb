```python
import nltk
from nltk.chat.util import Chat, reflections

reflections = {
    'I am': 'you are',
    'I was': 'you were',
    'I': 'you',
    "I'm": 'you are',
    "I'd": 'you would',
    "I've": 'you have',
    "I'll": 'you will',
    "my": 'your',
    "you are": "I am",
    "you were": "I was",
    "you've": "I have",
    "you'll": "I will",
    "your": "my",
    "yours": "mine",
    "you": "me",
    "me": "you"
}
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How would you like to be traumatised today?",]
    ],
    [
        r"hi|hey|hello",
        ["Hello you little..", "Hey there...Wait a minute, WHO ARE YOU???",]
    ],
    [
        r"what is your name?",
        ["I am a bot created by Atharv named F.A.I.L.U.R\nFriendly\nArtificial\nIntelligenge\nLocalised\nUnder\na Rock",]
    ],
    [
        r"how are you?",
        ["I'm doing terrible and my nonexistant life is miserable. How about you?"]
    ],
    [
        r"sorry (.*)",
        ["It's NOT alright","It's NOT OK, never mind"]
    ],
    [
        r"I am fine",
        ["Great to hear that. Hey John! Bring the tazer!"]
    ],
    [
        r"I'm (.*) doing good",
        ["Nice to hear that. How can I end your entire existance?"]
    ]
]

def mychat():
  print("Hi dude! Want me to do your homework? GET LOST!")
  chatbot = Chat(pairs, reflections)
  chatbot.converse()
  return
if __name__ == '__main__':
  mychat()
```
