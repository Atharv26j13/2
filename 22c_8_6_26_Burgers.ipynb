```python
import random

menu = ['Burger','Fries','Nuggets','Puffs','Sauce']
orders = []
for i in range(int(input("Enter the number of customers"))):
  if random.randint(1,3) == 3:
    if random.randint(1,2) == 2:
      orders.append([menu[random.randint(0,2)], menu[3], menu[4]])
    else:
      orders.append([menu[0], menu[1], menu[4]])

  elif random.randint(1,3) == 3:
    if random.randint(1,2) == 2:
      orders.append([menu[random.randint(0,2)], menu[3], menu[4], menu[1], menu[2]])
    else:
      orders.append([menu[0], menu[0], menu[3], menu[4]])

  else:
    if random.randint(1,2) == 2:
      orders.append([menu[random.randint(0,2)], menu[3], menu[4], menu[1], menu[2]])
    else:
      orders.append([menu[0]])

print(orders)
F=0
B=0
N=0


BurgerS = 0
FriesS = 0
NuggetsS = 0

BurgerF = 0

NuggetsF = 0

BurgerN = 0
FriesN = 0

for i in orders:
  # Reset counts for each order to check for presence in *this* order
  current_B_present = False
  current_F_present = False
  current_N_present = False
  has_sauce = False

  for x in i:
    if x == 'Burger':
      B += 1 # Accumulate total Burger count
      current_B_present = True
    if x == 'Fries':
      F += 1 # Accumulate total Fries count
      current_F_present = True
    if x == 'Nuggets':
      N += 1 # Accumulate total Nuggets count
      current_N_present = True
    if x == 'Sauce':
      has_sauce = True

  if current_B_present:
    if current_F_present:
      BurgerF +=1
    if current_N_present:
      BurgerN +=1
    if has_sauce:
      BurgerS +=1
  if current_F_present:
    if current_N_present:
      FriesN +=1
    if has_sauce:
      FriesS +=1
  if current_N_present:
    if current_F_present:
      NuggetsF+=1
    if has_sauce:
      NuggetsS+=1

print('Burgers:',B)
print('Fries:',F)
print('Nuggets:',N)
print('Burger and sauce:',BurgerS)
print('Burger and nuggets:',BurgerN)
print('Fries and sauce:',FriesS)
print('Fries and nuggets:',FriesN)
print('Nuggets and sauce:',NuggetsS)
print('Burger and fries:',BurgerF)
print('Burger and nuggets:',BurgerN)
print('Fries and nuggets:',FriesN)
```

```python
print("Welcome to Temu Popeyes!")
if B>F and B>N:
  a = input("Would you like to start with a Burger?: ")
  if BurgerS>BurgerF and BurgerS>BurgerN:
    b = input("Would you like to add sauce?: ")
  if BurgerF>BurgerS and BurgerF>BurgerN:
    b = input("Would you like to add fries?: ")
  if BurgerN>BurgerS and BurgerN>BurgerF:
    b = input("Would you like to add nuggets?: ")
if F>B and F>N:
  a = input("Would you like to start with Fries?: ")
  if FriesS>FriesN:
    b = input("Would you like to add sauce?: ")
  if FriesN>FriesS:
    b = input("Would you like to add burgers?: ")
if N>B and N>F:
  a = input("Would you like to start with Nuggets?:")
  if NuggetsS>NuggetsF:
    b = input("Would you like to add sauce?: ")
  if NuggetsF>NuggetsS:
    b = input("Would you like to add fries?: ")
c = input("Would you like to add anything else?: ")
print("Processing...")
print(f"Your order will be given to you of 1 {a}, 1 {c} and, 1{b}")
```
