```python
import random as rand
import matplotlib.pyplot as plt

# Initialize
a = []
b = []

a.append(rand.uniform(-2, 2))
for i in range(1, 15):
    a.append(a[i-1] + rand.uniform(-0.4, 0.2))

b.append(rand.uniform(-2, 2))
for i in range(1, 15):
    b.append(b[i-1] + rand.uniform(-0.2, 0.4))

print(a, "\n", b)

# Predict (differences)
preda = []
predb = []

for i in range(1, 15):
    preda.append(a[i] - a[i-1])
    predb.append(b[i] - b[i-1])

print(preda)
print(predb)

# Average change
predax = sum(preda) / len(preda)
predbx = sum(predb) / len(predb)

print(predax, predbx)

# Next actual step
a.append(a[-1] + rand.uniform(-0.4, 0.2))
b.append(b[-1] + rand.uniform(-0.2, 0.4))

# Prediction error
print(a[-1] - (a[-2] + predax))
print(b[-1] - (b[-2] + predbx))

# Plot
plt.scatter(b[:-1], a[:-1])
plt.plot(b[:-1], a[:-1])
plt.xlabel("b")
plt.ylabel("a")
plt.show()
```
