```python
import matplotlib.pyplot as plt
import random
```

```python
class Customer():
  def __init__ (self,salary,spending):
    self.salary = salary
    self.spending = spending
Clients = []
for i in range(100):
  Clients.append(Customer(random.randint(1,100),random.randint(1,100)))
plt.scatter([i.salary for i in Clients],[i.spending for i in Clients],color='black',s=60)
l1 = []
l2 = []
l3 = []
l4 = []
MaxSal = 0
MinSal = 100
MinSpend = 100
MaxSpend = 0
for i in Clients:
  if i.salary < 51 and i.spending < 51:
    l1.append(i)
  elif i.salary < 51 and i.spending > 50:
    l2.append(i)
  elif i.salary > 50 and i.spending < 51:
    l3.append(i)
  else:
    l4.append(i)
  if i.salary < MinSal:
    MinSal = i.salary
  if i.salary > MaxSal:
    MaxSal = i.salary
  if i.spending < MinSpend:
    MinSpend = i.spending
  if i.spending > MaxSpend:
    MaxSpend = i.spending
print(MinSal,MaxSal)
print(MinSpend,MaxSpend)
plt.scatter([i.salary for i in l1],[i.spending for i in l1])
plt.scatter([i.salary for i in l2],[i.spending for i in l2])
plt.scatter([i.salary for i in l3],[i.spending for i in l3])
plt.scatter([i.salary for i in l4],[i.spending for i in l4])
plt.scatter([MinSal],[1],s=70)
plt.scatter([MaxSal],[1],s=70)
plt.scatter([1],[MinSpend],s=70)
plt.scatter([1],[MaxSpend],s=70)
plt.xlabel('Salary')
plt.ylabel('Spending')
```
