```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.randint(1, 100)
y = 100 - x

print(x)
print(y)

sns.barplot(x=["x", "y"], y=[x, y])
plt.ylim(0, 100)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
VALUES = int(input("Enter amount of values:\n"))
xx = 0
yy = 0
x = []
y = []
for _ in range(VALUES):
  x.append(random.randint(random.randint(1,10),random.randint(40,50)))
  y.append(random.randint(random.randint(40,50),random.randint(90,100)))
for i in range(VALUES):
  xx += x[i]
  yy += y[i]
xx = xx/VALUES
yy = yy/VALUES
xxx = []
yyy = []
for i in range(VALUES):
  xxx.append(xx)
  yyy.append(yy)
line = []
for i in range(VALUES):
  line.append(xx+((yy-xx)/VALUES*i))
sns.lineplot(xxx,color="black")
sns.lineplot(yyy,color="black")
sns.lineplot(line,color="black")
plt.scatter(range(VALUES), x, color="blue", label="x values")
plt.scatter(range(VALUES), y, color="red", label="y values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
print("x:", x)
print("y:", y)
```
