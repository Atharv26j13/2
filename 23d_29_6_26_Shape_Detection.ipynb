```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread("shapes.png")
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
ret,thresh = cv2.threshold(gray_img,50,255,1)
cv2_imshow(thresh)
```

```python
contours,h = cv2.findContours(thresh,1,2)
```

```python
for cnt in contours:
  approx = cv2.approxPolyDP(cnt,0.01*cv2.arcLength(cnt,True),True)
  n = len(approx)
  if n==6:
    print("We hava a hexagon here")
    cv2.drawContours(img,[cnt],0,255,10)
  elif n==3:
    print("We have a triangle here")
    cv2.drawContours(img,[cnt],0,(0,255,0),3)
  elif n>9:
    print("We have a circle here")
    cv2.drawContours(img,[cnt],0,(0,255,255),3)
  elif n==4:
    print("We have a square here")
    cv2.drawContours(img,[cnt],0,(255,0,0),3)
cv2_imshow(img)
```
