```python
import random

l = []

class WATCH:
  def __init__(self,name,mos,genre,length,yearor,rating):
    self.name=name
    self.mos=mos
    self.genre=genre
    self.length=length
    self.yearor=yearor
    self.rating=rating
    l.append(self)

CONJURING = WATCH("CONJURING","Movie","Horror",120,2015,9)
CONJURING2 = WATCH("CONJURING 2","Movie","Horror",120,2016,7)
DUNE = WATCH("DUNE","Movie","Sci-Fi",150,2022,9.5)
DUNE2 = WATCH("DUNE 2","Movie","Sci-Fi",180,2025,8)
F1 = WATCH("F1","Movie","Racing",170,2025,9.5)
FASTANDFURIOUS = WATCH("FAST AND FURIOUS","Movie","Action",120,2017,9)
FASTANDFURIOUS2 = WATCH("FAST AND FURIOUS 2","Movie","Action",100,2019,8)
FASTANDFURIOUS3 = WATCH("FAST AND FURIOUS 3","Movie","Action",120,2020,9)
FASTANDFURIOUS4 = WATCH("FAST AND FURIOUS 4","Movie","Action",140,2022,7)
SONIC = WATCH("SONIC","Movie","Action",100,2020,7)
SONIC2 = WATCH("SONIC 2","Movie","Action",120,2022,8)
SONIC3 = WATCH("SONIC 3","Movie","Action",150,2025,9.5)
JJK0 = WATCH("Jujutsu Kaisen 0","Movie","Anime",120,2020,8)
JJK = WATCH("Jujutsu Kaisen","Show","Anime",3,2019,9)
NARUTO = WATCH("Naruto","Show","Anime",9,2015,8.5)
POP = WATCH("Prince of Peoria","Show","Comedy",2,2017,7)
TEENTITANSGO = WATCH("Teen Titans Go","Show","Comedy",5,2015,7)
GOKU = WATCH("Naruto","Show","Anime",9,2015,8.5)
MINECRAFTMOVIE = WATCH("Minecraft","Movie","Adventure",100,2025,7)
FNAF1 = WATCH("FNAF 1","Movie","Horror",120,2023,7)
FNAF2 = WATCH("FNAF 2","Movie","Horror",120,2026,8.5)
ANNABELLE = WATCH("Annabelle","Movie","Horror",120,2023,9.5)
ISITCAKE = WATCH("Is it a cake?","Show","Comedy",4,2020,9)
NINJAGO = WATCH("Lego Ninjago","Show","Action",11,2013,9.5)
FULLERHOUSE = WATCH("Fuller House","Show","Comedy",5,2020,9)
BENHUR = WATCH("Ben Hur","Movie","Action",230,1970,9.5)
GLADIATOR = WATCH("Gladiator","Movie","Action",150,2000,9.5)

def p3():
  x = 0
  a = input("\nSEARCH:")
  print("Results:")
  for i in l:
    if (a.lower() in i.name.lower() or
        a.lower() in i.genre.lower()):
      print(i.name)
      x = 1

  if x == 0:
    print("NO RESULTS FOUND\nTRY AGAIN:\n")
    p3()


def p2():
  x = 0
  a = int(input("By what do you want to search?\n1.Name\n2.Genre\n3.Length\n4.Rating\n"))

  if a == 1:
    b = input("\nENTER NAME:")
    print("Results:")
    for i in l:
      if b.lower() in i.name.lower():
        print(i.name)
        x = 1

  elif a == 2:
    b = input("\nENTER GENRE:")
    print("Results:")
    for i in l:
      if b.lower() in i.genre.lower():
        print(i.name)
        x = 1

  elif a == 3:
    b = int(input("\nENTER MINIMUM LENGTH:"))
    c = int(input("\nENTER MAXIMUM LENGTH:"))
    print("Results:")
    for i in l:
      if i.length >= b and i.length <= c:
        print(i.name)
        x = 1

  elif a == 4:
    b = float(input("\nENTER RATING:"))
    print("Results:")
    for i in l:
      if i.rating == b:
        print(i.name)
        x = 1

  else:
    print("INVALID INPUT")
    p2()
    return

  if x == 0:
    print("NO RESULTS FOUND\nTRY AGAIN:\n")
    p2()


# Genre lists
lor = []
lor1 = []
lor2 = []
lor3 = []
lor4 = []
lor5 = []
lor6 = []
lor7 = []

for i in l:

  if i.genre == "Horror":
    lor1.append(i)
  elif i.genre == "Sci-Fi":
    lor2.append(i)
  elif i.genre == "Racing":
    lor3.append(i)
  elif i.genre == "Action":
    lor4.append(i)
  elif i.genre == "Anime":
    lor5.append(i)
  elif i.genre == "Adventure":
    lor6.append(i)
  elif i.genre == "Comedy":
    lor7.append(i)


def p4():
  if len(lor) == 0:
    print("SEARCH FOR A FEW MOVIES OR SHOWS SO WE CAN RECOMMEND YOU SOMETHING")
  else:
    print("RECOMMENDATIONS:")

    def recommend(genre):
      if genre == "Horror" and lor1:
        print("Horror:", random.choice(lor1).name)
        for i in lor1:
          if i.rating >= 9:
            print("Horror:",i.name)
      elif genre == "Sci-Fi" and lor2:
        print("Sci-Fi:", random.choice(lor2).name)
        for i in lor2:
          if i.rating >= 9:
            print("Sci-Fi:",i.name)
      elif genre == "Racing" and lor3:
        print("Racing:", random.choice(lor3).name)
        for i in lor3:
          if i.rating >= 9:
            print("Racing:",i.name)
      elif genre == "Action" and lor4:
        print("Action:", random.choice(lor4).name)
        for i in lor4:
          if i.rating >= 9:
            print("Action:",i.name)
      elif genre == "Anime" and lor5:
        print("Anime:", random.choice(lor5).name)
        for i in lor5:
          if i.rating >= 9:
            print("Anime:",i.name)
      elif genre == "Adventure" and lor6:
        print("Adventure:", random.choice(lor6).name)
        for i in lor6:
          if i.rating >= 9:
            print("Aventure:",i.name)
      elif genre == "Comedy" and lor7:
        print("Comedy:", random.choice(lor7).name)
        for i in lor7:
          if i.rating >= 9:
            print("Comedy:",i.name)

    recommend(lor[-1])

    if len(lor) > 1:
      recommend(lor[-2])

    if len(lor) > 2:
      recommend(lor[-3])


def NETFLIX():
  print("Welcome to NETFLIX")
  print("What would you like to watch?")
  a = int(input("1:Most Watched\n2:Find\n3:Search\n4:Recommended\n"))
  x = 0

  if a == 1:
    print("\nMost Watched:")
    print("1.CONJURING")
    print("2.DUNE")
    print("3.F1")
    print("4.FASTANDFURIOUS")
    print("5.SONIC")

  elif a == 3:
    p3()

  elif a == 2:
    p2()

  elif a == 4:
    p4()

  a = input("\nWHICH ONE WOULD YOU LIKE TO WATCH?:")
  for i in l:
    if a.lower() == i.name.lower():
      print("\nName:", i.name)
      print("Type:", i.mos)
      print("Genre:", i.genre)
      print("Year:", i.yearor)
      if i.mos == "Movie":
        print("Minutes:", i.length)
      else:
        print("Seasons:", i.length)
      x = 1
      lor.append(i.genre)

  if x == 0:
    print("NONE FOUND BY THIS NAME")

  input("\nPress enter to go home:")
  NETFLIX()


NETFLIX()
```
