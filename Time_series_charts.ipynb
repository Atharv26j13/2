```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = pd.read_csv('gapminder(2007).csv')
data.head()
```

```python
plt.pie(data.groupby('continent').size(),autopct='%.2f',labels=data.groupby('continent').size().index,labeldistance=1.15,wedgeprops={'linewidth':2,'edgecolor':'white'})
```

```python
data_group = data.groupby('continent')['population'].mean()

data_group.plot.area()

plt.ylabel('Average Population')
plt.xlabel('Continent')
plt.show()
```
