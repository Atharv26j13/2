```python
import sqlite3
database = 'database (3).sqlite'
conn = sqlite3.connect(database)
print('Opened data successfully')
```

```python
import pandas as pd
table = pd.read_sql("""SELECT * FROM sqlite_master WHERE type = 'table'""",conn)
table
```

```python
team = pd.read_sql("""SELECT * FROM team""",conn)
team
```

```python
MI_wins = pd.read_sql("""SELECT * FROM Match WHERE Match_Winner == 7;""",conn)
MI_wins
```

```python
MI_winsS = pd.read_sql("""SELECT * FROM Match WHERE Match_Winner == 7 and Season_Id IN (8,9);""",conn)
MI_winsS
```

```python
MI_winsM = pd.read_sql("""SELECT * FROM Team WHERE Team_Name LIKE 'De%';""",conn)
MI_winsM
```

```python
MI_winsMX = pd.read_sql("""SELECT MIN(Win_Margin), MAX(Win_Margin) FROM Match;""",conn)
MI_winsMX
```
