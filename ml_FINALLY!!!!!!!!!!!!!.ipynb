```python
#required special function
print(__doc__)
#import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
#load diabetes dataset into X and Y for graph
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
#Train and test variable for X side of graph
diabetes_X = diabetes_X[:,np.newaxis,2]
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
#Train and test variable for Y side of graph
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]
#Create variable to do the linear regression
regr = linear_model.LinearRegression()
#Based on the trained data, fitting patterns using regression
regr.fit(diabetes_X_train,diabetes_y_train)
#Based on patterns, predicting for the X side of the graph
diabetes_y_pred = regr.predict(diabetes_X_test)
#Printing coefficients and error based on prediction
print(regr.coef_)
print(mean_squared_error(diabetes_y_test,diabetes_y_pred))
#Create scatterplot and line plot
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.plot(diabetes_X_test, diabetes_y_pred,color='blue',linewidth=3)
#Plot scatterplot and line plot
plt.xticks(())
plt.yticks(())
plt.show()
#X = the data given over time
#Y = avg growth slope
```
