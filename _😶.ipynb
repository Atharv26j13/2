```python
data = pd.read_csv("Titanic Dataset.csv")
data.head()
```

```python
data.dtypes
```

```python
nom_cat = ['Name','Ticket','Cabin']
ord_cat = ['Embarked','Gender']
data['Gender'] = pd.Categorical(data['Gender'],categories=['male','female'])
data
```

```python
median_index = np.median(data['Gender'].cat.codes)
median_gender = data['Gender'][int(median_index)]
print(median_gender)
```

```python
data['Embarked'].value_counts()
```

```python
embarked_categories = ['S','C','Q']
data['Embarked'] = pd.Categorical(data['Embarked'],categories=embarked_categories, ordered=True)
median_index = np.median(data['Embarked'].cat.codes)
median_embarked = data['Embarked'][int(median_index)]
print(median_embarked)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=data)
```

```python
sns.countplot(x='Gender',hue='Survived',data=data)
```

```python
sns.countplot(x='Survived',data=data, palette='summer')
```
